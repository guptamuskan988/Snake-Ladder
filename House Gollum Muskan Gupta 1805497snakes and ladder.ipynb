```python
import numpy as np
import pandas as pd
import random
print("#####Welcome To Snake and Ladder Game#####")
player1=input("Enter the name of player1 ")
player2=input("Enter the name of player2 ")
player1=player1.upper()
player2=player2.upper()
s={17:7,54:34,62:19,98:79}
l={3:38,24:33,42:93,72:84}
pos=[0,0,0]#pos[0]=0 so that i can easily refer to players using 1 & 2
chance=0
mode=[0,0,0]#to decide auto mode or manual for auto=1 and manual =0
def automode(k):
    no=random.randint(1,6)
    print("You got ",no)
    pos[k]=pos[k]+no;
    for i in s:
        if i==pos[k]:
            pos[k]=s[i]
            print("OOPS! Snake")
    for j in l:
        if j==pos[k]:
            pos[k]=l[j]
            print("YAYY! Ladder")
    if pos[k]>100:
        pos[k]=pos[k]-no
    print("Your Final position is:",pos[k])
    return
def manualmode(no,k):
    if no>0 and no<21:
        pos[k]=pos[k]+no;
        for i in s:
            if i==pos[k]:
                pos[k]=s[i]
                print("OOPS! Snake")
        for j in l:
             if j==pos[k]:
                pos[k]=l[j]
                print("YAYY! Ladder")
    if pos[k]>100:
        pos[k]=pos[k]-no
    print("Your Final position is:",pos[k])
    return  
#calculation of position based on mode
def calculate(k):
    z='play'
    if mode[k]==1:
        res=input("Enter Roll for auto generation of number ")
        res=res.lower()
        while res!='roll':
            if res=='quit':
                z=res
                break
            else:
                res=input("Please enter Roll for auto generation of number ")
                res=res.lower() 
        if res=='roll':
            automode(k)
        else:
            return z
    else:
        str=input("Enter a no between 1-20: ")
        if str=='quit':
            z=str
            return z
        else:
            no=int(str)
            while no<=0 or no>=21:
                no=input("Please input a no between 1-20: ")
                if no=='quit':
                    z=no
                    break
                else:
                    no=int(no)
            if no=='quit':
                return z
            else:
                manualmode(no,k)
#asking players about auto or manual mode               
str=input("Player 1: In which mode you want to enter number Auto or Manual: ")
str=str.lower()
if str=='auto':
    mode[1]=1
elif str=='manual':
    mode[1]=0
else:
    while str!='auto' and str!='manual':
        str=input("Player 1: Please Enter correct mode Auto or Manual: ")
        str=str.lower()
    if str=='auto':
        mode[1]=1
    else:
        mode[1]=0
str=input("Player 2:In which mode you want to enter number Auto or Manual: ")
str=str.lower()
if str=='auto':
    mode[2]=1
elif str=='manual':
    mode[2]=0
else:
    while str!='auto' and str!='manual':
        str=input("Player 2: Please Enter correct mode Auto or Manual else : ")
        str=str.lower()
    if str=='auto':
        mode[2]=1
    else:
        mode[2]=0
#Playing game
while pos[1]!=100 and pos[2]!=100:
    chance=chance+1
    if str=='quit':
        break
    elif chance%2==0:
        print("Player2: ",player2," chance")
        k=2
        str=calculate(k)
    else:
        print("Player1: ",player1,"chance")
        k=1
        str=calculate(k)
if(pos[1]==100):
    print("Player 1",player1," won the game")
    print("#####Game Successfully Finished####")
elif(pos[2]==100):
    print("Player 2",player2," won the game")
    print("#####Game Successfully Finished####")
else:
    if k==1:
        print("Player:",player2,"Won the game")
        print("#####Game Successfully Finished####")
    else:
        print("Player:",player1," Won the game")
        print("#####Game Successfully Finished####")

```
